# Visualizing Loans Awarded by Kiva

The dataset contains information about loans awarded by the non-profit <a href = "https://www.kiva.org/" target = "_blank">Kiva</a>. 

Using Seaborn, I will explore the average loan amount by country using aggregated bar charts. I will also visualize the distribution of loan amounts by project type and gender using box plots and violin plots.

## Importing Necessary Python Modules

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

## Ingesting The Data
Loaded **kiva_data.csv** into a DataFrame called `df`. 

```python
df = pd.read_csv('kiva_data.csv')
df.head()
```

## Examining The Data

### Overview of the dataset:

Each entry (row) in the dataset represents a loan that Kiva awarded to a particular project. The `loan_amount` column shows the amount (in U.S. dollars) awarded to the project. The `activity` column has the category type that the project falls under. The `country` column is the country where the project is located. The `gender` column represents the gender of the primary person who applied for the loan. 

```python
df.head(25)
```

## Bar Charts

Created a bar plot using Seaborn to visualize the average size of Kiva loans given to projects, by country.

```python
# Creates the figure, note: you're only using this syntax so that you can modify the y-axis ticks later
f, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x = 'country', y = 'loan_amount', data = df)
plt.xlabel('Country')
plt.ylabel('Loan Amount (USD)')
```

### Adding `$` units

Setting `loan_amount` ticks on the y-axis begin with a `$` (units of USD). 

```python
import matplotlib.ticker as mtick

# Creating the figure
f, ax = plt.subplots(figsize=(15, 10))

# Plotting the data
sns.barplot(data=df, x="country", y = "loan_amount", hue = 'activity')

# Formating the y-axis
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

# Axis labels and title
plt.xlabel('Country')
plt.ylabel('Loan Amount (USD)')
plt.title('Kiva Donations per Country')

```

## Learing More By Using `hue` 

To extract further insights, I choose to visualize the loan amount by country, and "nesting" by gender. 

```python
# Creating the figure
f, ax = plt.subplots(figsize=(15, 10))

# Plotting the data
sns.barplot(data=df, x="country", y = "loan_amount", hue = 'gender')

# Formating the y-axis
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

# Axis labels and title
plt.xlabel('Country')
plt.ylabel('Loan Amount (USD)')
plt.title('Kiva Donations per Country')
```

## Styling

```python
# Set color palette
sns.set_palette('Pastel1')

# Set style
sns.set_style('white')
sns.set_context('paper', font_scale = 1.5)

# Create figure and axes
plt.figure(figsize=(15, 10))
ax = plt.subplot()
sns.barplot(x = 'gender', y = 'loan_amount', data = df)

# Formatting the y-axis
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

# Axis labels and title
ax.set_title('Loan Amounts by Gender')
plt.xlabel('Gender')
plt.ylabel('Loan Amounts')
```

## Box Plots

So far I have visualized the average size of loans by country using bar charts; now I will continue by making a box plot to compare the distribution of loans by country.

```python
# Styling
sns.set_palette('Set1')

# Plotting
plt.figure(figsize=(16, 10))
sns.boxplot(x = 'country', y = 'loan_amount', data = df)

# Formatting the y-axis
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

# Axis labels and title
ax.set_title('Loan Amounts by Country')
plt.xlabel('Country')
plt.ylabel('Loan Amounts')
```

#### Reflection Questions

Which country's box has the widest distribution?

        Kenya has the widest distribution 

In which country would you be most likely to receive the largest loan amount?

        According to the median values provided by the boxplot, Cambodia is most likely to recieve the 
        largest loan amount. 

## Box Plot by Activity

```python
# Styling
sns.set_palette('Set2')
sns.set_style('white')

# Boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x = 'activity', y = 'loan_amount', data = df)

# Axis labels and title
plt.title('Breakdown of Loans by Activities')
plt.ylabel('Loan Amount (USD)')
plt.xlabel('Activity')
```

#### Reflection Questions

What does this visualization reveal that previous ones did not?

        This visualization provides greater information by breaking down the loan amounts according to activity
        types. 
        
        The median loan awarded for farming activities is higher relative to food production/sales and 
        general store.
        
        Furthermore, farming has the widest distribution

## Violin Plots

To gather further insights I will switch over to a violin plot to gain a better understanding of our dataset's
distribution

```python
# Styling
sns.set_palette('Pastel1')
sns.set_style('white')

# Plotting the violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(data=df, x="activity", y="loan_amount")

# Axis labels and titles 
plt.title('Breakdown of Loans by Activities')
plt.ylabel('Loan Amount (USD)')
plt.xlabel('Activity')
```

### Distribution of loan amount by country.


```python
# Styling
sns.set_palette('Set2')
sns.set_style('white')

# Plotting the violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(data=df, x="country", y="loan_amount")

# Axis labels and titles 
plt.title('Breakdown of Loans by Country')
plt.ylabel('Loan Amount (USD)')
plt.xlabel('Country')
```

## Split Violin Plots

```python
# Styling
sns.set_palette("Spectral")
sns.set_style('white')

# Plotting the violin graph
plt.figure(figsize=(18, 12))
sns.violinplot(x = 'country', y = 'loan_amount', data = df, hue = 'gender', split = True)

# Axis labels and titles 
plt.title('Breakdown of Loans by Country')
plt.ylabel('Loan Amount (USD)')
plt.xlabel('Country')
```

#### Reflection Questions

What does this visualization reveal about the distribution of loan amounts within countries by gender?
